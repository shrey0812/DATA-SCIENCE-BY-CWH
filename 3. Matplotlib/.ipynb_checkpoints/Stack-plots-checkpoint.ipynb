```python
import matplotlib.pyplot as plt
```

```python
activities = ['Studying','Playing','Watching TV','Sleeping']
```

```python
time_spent = [3,2,2,5]
```

```python
colors = ['blue','red','yellow','pink']
```

```python
plt.figure(figsize=(6,6))
```

```python
plt.pie(time_spent,labels=activities,colors=colors,startangle=90,autopct='%1.1f%%')
plt.title('School going students')
plt.axis('equal')
```

```python
days = [1, 2, 3, 4, 5, 6, 7]  # Days of the week
studying = [3, 4, 3, 5, 4, 3, 4]
playing = [2, 2, 1, 1, 2, 3, 2]
watching_tv = [2, 1, 2, 2, 1, 1, 1]
sleeping = [5, 5, 6, 5, 6, 5, 5]

plt.stackplot(days,sleeping,playing,watching_tv,sleeping,labels=activities,colors=colors,alpha = 0.4)
plt.legend(loc = 'upper left')
plt.title("Students time distribution over activities during the week")
plt.xlabel("days")
plt.ylabel("hours")
plt.grid(True)
```
