```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1,2,3,4,5,6])
y1 = x*2
y2 = x**2
```

```python
y1
```

```python
y2
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1,label='Multiply by 2')
plt.title("Changes by multiplying with 2")
plt.legend()

plt.subplot(1,2,2)
plt.plot(x,y2,label='Power of 2')
plt.title("Changes by squaring with 2")
plt.legend()
```

```python
x = np.arange(100)
y1 = x**2
y2 = np.power(x,3)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1,label='Power by 2')
plt.title("Changes by power of 2")
plt.xlabel("Number")
plt.ylabel("Number Squared")
plt.legend()

plt.subplot(1,2,2)
plt.plot(x,y2,label='Power of 3')
plt.title("Changes by power of 3")
plt.xlabel("Number")
plt.ylabel("Number Cubed")
plt.legend()
plt.tight_layout()
```

```python
x = np.arange(100)
y1 = x**2
y2 = np.power(x,3)
y3 = np.sqrt(x)
y4 = np.power(x,(1//3))
```

```python
plt.subplot(2,2,1)
plt.plot(x,y1,label='Power by 2')
plt.title("Changes by power of 2")
plt.xlabel("Number")
plt.ylabel("Number Squared")
plt.legend()

plt.subplot(2,2,2)
plt.plot(x,y2,label='Power of 3')
plt.title("Changes by power of 3")
plt.xlabel("Number")
plt.ylabel("Number Cubed")
plt.legend()

plt.subplot(2,2,3)
plt.plot(x,y1,label='Square Root')
plt.title("Changes by Square Root")
plt.xlabel("Number")
plt.ylabel("Square Root of Number")
plt.legend()

plt.subplot(2,2,4)
plt.plot(x,y2,label='Cube Root')
plt.title("Changes by Cube Root")
plt.xlabel("Number")
plt.ylabel("Cube Root of Number")
plt.legend()
plt.tight_layout()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(8, 4))
 
axs[0].plot(x, y1)
axs[0].set_title('x * 2')
 
axs[1].plot(x, y2)
axs[1].set_title('x squared')
 
fig.suptitle('Simple Comparison Plots', fontsize=14)
fig.tight_layout()
fig.subplots_adjust(top=0.85)
```
