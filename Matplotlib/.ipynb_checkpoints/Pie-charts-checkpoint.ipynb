```python
import matplotlib.pyplot as plt
```

```python
plt.style.use('ggplot')
```

```python
labels = ["Sachin","Sehwag","Kohli","Yuvraj"]
runs = [18000,8000,12000,9500]

plt.title("Career Runs of Indian Batsmen")
plt.pie(runs,labels=labels)
```

```python
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
plt.pie(runs, labels=labels, colors=colors)
```

```python
plt.pie(
    runs,
    labels=labels,
    colors=colors,
    wedgeprops={'edgecolor':'black','linewidth':2,'linestyle':'--'}
)
```

```python
explode = [0.1,0,0,0] # only highlight sachin's scale
plt.pie(runs,labels=labels,colors=colors,explode=explode)
plt.title("Exploded Pie example")
```

```python
plt.pie(runs,labels=labels,colors=colors,explode=explode,shadow=True)
```

```python
plt.pie(runs,labels=labels,colors=colors,explode=explode,shadow=True,startangle=150)
```

```python
plt.pie(runs,labels=labels,colors=colors,explode=explode,shadow=True,startangle=150,autopct='%1.1f%%')
```

```python
# Too many categories Bad Pie chart need to avoid
languages = ['Python', 'Java', 'C++', 'JavaScript', 'C#', 'Ruby', 'Go', 'Rust', 'Swift', 'PHP']
usage = [30, 20, 10, 10, 7, 5, 4, 3, 2, 1]

plt.pie(usage, labels=languages, autopct='%1.1f%%', startangle=90)
plt.title("Programming Language Usage (Crowded Example)")
plt.show()
```
